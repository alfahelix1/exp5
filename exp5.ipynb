```python
'''Problem 1
A dictionary stores product names and their prices.
Write a python program to update the price of a given product.'''
# Create a dictionary to store product names and their prices
products = {
    "Laptop": 1200,
    "Mouse": 25,
    "Keyboard": 75,
}
print("Original products dictionary:", products)


product_to_update = "Mouse"
new_price = 30

# Check if the product exists before updating
if product_to_update in products:
    products[product_to_update] = new_price
    print(f"Updated the price of {product_to_update} to ${new_price}.")
else:
    print(f"{product_to_update} not found in the dictionary.")

print("Products dictionary after update:", products)
```

```python
'''Program 2
A dictionary stores student names and marks
Write a python to search for a students marks using the name entered by the user
If the student is not found, display a suitable message.'''
students = {
    "Dhrisha": 85,
    "Bhide": 92,
    "Chintu": 78,
}

student_name = input("Enter the student's name to search:")

if student_name in students:
    marks = students[student_name]
    print(f"{student_name}'s marks: {marks}")
else:
    print(f"Student '{student_name}' not found in the dictionary.")

```

```python
'''Program 3
write a python program to validate user login.'''
#Create a dictionary to store usernames and passwords
users = {
    "admin": "password123",
    "user1": "securepass",
    "guest": "guestpass"
}

# Get username and password from the user
username = input("Enter username: ")
password = input("Enter password: ")

# Validate login
if username in users:
    if users[username] == password:
        print(f"Login successful! Welcome, {username}.")
    else:
        print("Invalid password.")
else:
    print("Username not found.")
```

```python
'''4. A dictionary stores student names and marks
Write a Python program to find the student who scored the highest marks.'''
#Create a dictionary to store student names and marks
marks = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
}
topper = max(marks, key=marks.get)
print("Topper:", topper)
print("Marks:", marks[topper])

```
